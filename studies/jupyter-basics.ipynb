#### jupyter command mode
- change cell type
    - code : Y
    - markdown : M
- move between cells
    - above : up arrow
    - below : down arrow
- create cell
    - above : A
    - below : B
- delete cell : D (double-tap)
    - undo : Z

#### coding in jupyter
- function description pop-up : shift + tab

_Detailed keybinds are available in Help > Keyboard Shortcuts_
