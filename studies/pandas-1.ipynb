#### Series
- a _Series_ is a one-dimensional array that holds any data type
- axis labels are referred as the _index_
- common data types include integers, floats and strings

- [Series API reference](https://pandas.pydata.org/docs/reference/api/pandas.Series.html) : attributes, methods etc



```python
number_list = [2022, 2018, 2014, 2010, 2006, 2002]
```

```python
# outputs the series along with its data type
pd.Series(number_list)
```

```python
string_list = ['cereal', 'milk', 'eggs', 'bread', 'fruit']
```

```python
# a series of strings corresponds to the object data type
pd.Series(string_list)
```

```python
# series + index creation

# length of both lists needs to be the same
prev_wc_win = ['Argentina', 'France', 'Germany', 'Spain', 'Italy', 'Brazil']
prev_wc_events = [2022, 2018, 2014, 2010, 2006, 2002]

# attach it to a variable so we can use it later
prev_wc_series = pd.Series(prev_wc_win, index=prev_wc_events, name='prev wc champs')

prev_wc_series
```

```python
# indexes can be basically anything

# what if the events were the values
inverted_prev_wc_series = pd.Series(prev_wc_events, index=prev_wc_win, name='prev wc events')

inverted_prev_wc_series # series data type changes to int
```

```python
# data type averages to object
some_data = {'name': 'Luigi', 'age':26, 'work': 'unemployed'}

pd.Series(some_data)
```

```python
# index values need to be the same as dictionary keys, else it returns NaN

# misspelling 'work' as 'job' causes the value to be not a number
pd.Series(some_data, index=['name', 'age', 'job'])
```

- attributes often give data about the object, while methods perform actions on that object, changing the object

```python
# reviving wc series 
print(inverted_prev_wc_series)
print(prev_wc_series)
```
